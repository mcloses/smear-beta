```python
from pathlib import Path

from ultralytics import YOLO
from IPython.display import display, Image

from src.utils.config import load_config
```

```python
import torch
torch.cuda.empty_cache()
```

CONFIG

```python
global_config, local_config = load_config()

pre_trained_model_filename = "yolov8m-seg.pt"
pre_trained_model_path = local_config["PATH"]["yolo_pretrained_models"]+pre_trained_model_filename
dataset = local_config["PATH"]["yolo_routes_data"]+local_config["FILENAMES"]["yolo_dataset_file"]
```

```python
epochs = 50
batch = 2
imgsz = 640
device = 0 #CUDA
project_name = "hold-detector"
verbose = True
single_cls = True
```

TRAINING

```python
model = YOLO(pre_trained_model_path)
```

```python
model.train(data=dataset, epochs=epochs, batch=batch, imgsz=imgsz, verbose=verbose, workers=0, single_cls=single_cls)
```

```python
path = local_config["PATH"]["models"]+"instance_segmentation/yolo/first_test.pt"
model.export()
```
